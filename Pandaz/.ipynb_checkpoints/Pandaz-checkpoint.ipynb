```python
import finpy_tse as fpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
max_99 = pd.read_csv('max_99.csv')
max_99
```

```python
def get_max_year(name):
    data = fpy.Get_Price_History(stock = name,
                                 start_date = '1402-01-01',
                                 end_date = '1402-08-28',
                                 ignore_date = False,
                                 adjust_price = True,
                                 show_weekday = False,
                                 double_date = False)
    maximum = max(data['Adj Close'])
    return data[data['Adj Close'] == maximum]
```

```python
get_max_year('شلیا')
```

###### Run Cell If necessary:

```python
max_year = pd.DataFrame()
count = 1
for index, row in max_99.iterrows():
    try:
        max_year = pd.concat([max_year,get_max_year(row['Ticker'])], ignore_index = True)
        print(str(count)+'- '+'Done!')
        count+=1
    except:
        print(str(count)+'- '+'Failed!')
        count+=1
max_year.to_csv('max_year.csv')
```

```python
max_year = pd.read_csv('max_year.csv')
```

```python
flag = 1
result = []
for index, row in max_99.iterrows():
    try:
        close_99 = np.int64(row['Adj Close'])
        close_year = np.array(max_year[max_year['Ticker'] == row['Ticker']]['Adj Close'])
        if ((close_99 * 2) + close_99) <= close_year[0] :
            print(str(flag)+'- '+row['Ticker'])
            result.append(row['Ticker'])
        else:
            print(str(flag)+'- NO!')
        flag += 1
    except Exception as error:
        print(error)
        flag += 1
```

```python
res = []
[res.append(i) for i in result if i not in res]
len(res)
```

```python
df = pd.DataFrame()
c = 1
for i in res:
    try:
        data = fpy.Get_Price_History(stock = i,
                                     start_date = '1402-01-01',
                                     end_date = '1402-08-28',
                                     ignore_date = False,
                                     adjust_price = True,
                                     show_weekday = False,
                                     double_date = True)
        df = pd.concat([df,data], ignore_index = False)
        print(str(c)+'- '+'Done!')
        c+=1
    except:
        print(str(c)+'- '+'Failed!')
        c+=1
df.to_csv('result_1402.csv')
```

##### Transform df to 2 dfs for price and volume

```python
names = ['پشاهن','سفانو','سپرمی','ثامان','بوعلی','وسالت','سیستم','کصدف','کورز','خعمرا','ولقمان','حرهشا','خفولا','خدیزل','فراز','سپیدار','دکپسول','وآتوس','کساپا','کازرو','بتک','شستان','خاور','خکاوه','کباده','اتکاسا','غالبر','ومشان','سمتاز','شسم','فجوش']
df_price = pd.DataFrame()
df_price['Date'] = pd.unique(df['J-Date']).tolist()
df_volume = pd.DataFrame()
df_volume['Date'] = pd.unique(df['J-Date']).tolist()

for i in names:
    close = []
    volume = []
    for index, row in df.iterrows():
        if row['Ticker']==i:
            close.append(np.int64(row['Adj Close']))
            volume.append(np.int64(row['Volume']))
    df_price[i] = close
    df_volume[i] = volume
```
