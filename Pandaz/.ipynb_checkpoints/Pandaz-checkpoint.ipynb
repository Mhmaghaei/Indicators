```python
import finpy_tse as fpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import jdatetime as jd
```

```python
max_99 = pd.read_csv('max_99.csv')
max_99
```

```python
today = jd.date.today().strftime("%Y-%m-%d")
'/History-'+ today + '.csv'
```

```python
def get_max_year(name):
    today = jd.date.today().strftime("%Y-%m-%d")
    data = fpy.Get_Price_History(stock = name,
                                 start_date = '1402-01-01',
                                 end_date = today,
                                 ignore_date = False,
                                 adjust_price = True,
                                 show_weekday = False,
                                 double_date = False)
    maximum = max(data['Adj Close'])
    return data[data['Adj Close'] == maximum]
```

###### Run Cell If necessary:

```python
max_year = pd.DataFrame()
count = 1
for index, row in max_99.iterrows():
    try:
        max_year = pd.concat([max_year,get_max_year(row['Ticker'])], ignore_index = True)
        print(str(count)+'- '+'Done!')
        count+=1
    except:
        print(str(count)+'- '+'Failed!')
        count+=1
max_year.to_csv('max_year.csv')
```

```python
max_year = pd.read_csv('max_year.csv')
```

```python
flag = 1
result = []
for index, row in max_99.iterrows():
    try:
        close_99 = np.int64(row['Adj Close'])
        close_year = np.array(max_year[max_year['Ticker'] == row['Ticker']]['Adj Close'])
        if ((close_99 * 2) + close_99) <= close_year[0] :
            print(str(flag)+'- '+row['Ticker'])
            result.append(row['Ticker'])
        else:
            print(str(flag)+'- NO!')
        flag += 1
    except Exception as error:
        print(error)
        flag += 1
```

```python
res = []
[res.append(i) for i in result if i not in res]
len(res)
```

```python
for i in res : print(i)
```

```python
df = pd.DataFrame()
c = 1
for i in res:
    try:
        data = fpy.Get_Price_History(stock = i,
                                     start_date = '1402-01-01',
                                     end_date = '1402-08-28',
                                     ignore_date = False,
                                     adjust_price = True,
                                     show_weekday = False,
                                     double_date = True)
        df = pd.concat([df,data], ignore_index = False)
        print(str(c)+'- '+'Done!')
        c+=1
    except:
        print(str(c)+'- '+'Failed!')
        c+=1
df.to_csv('result_1402.csv')
```
