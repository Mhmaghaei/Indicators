```python
import finpy_tse as fpy
import pandas as pd
```

```python
tickers = pd.read_csv('Tickers.csv')
names = []
for i in tickers['Ticker']:
    names.append(i)
```

```python
def get_max_99(name):
    data = fpy.Get_Price_History(stock = name,
                                 start_date = '1399-1-1',
                                 end_date = '1399-12-28',
                                 ignore_date = False,
                                 adjust_price = True,
                                 show_weekday = False,
                                 double_date = False)
    maximum = max(data['Adj Close'])
    return data[data['Adj Close'] == maximum]
```

```python
max_99 = pd.DataFrame()
count = 1
for name in names:
    try:
        max_99 = pd.concat([max_99,get_max_99(name)], ignore_index = True)
        print(str(count)+'- '+'Done!')
        count+=1
    except:
        print(str(count)+'- '+'Failed!')
        count+=1
max_99.to_csv('max_99.csv')
```
